## EDA
- Categorical variable distributions over time
    - Gives period level plot (Monthly, weekly, quarterly) of SKU/PPG based on user input
    - Outputs: Location: Output/EDA : 1. Categorical distribution plot. 2. Dataset with period level summary
- Chronic missing and total missing dates from the dataframe
    - Returns number of instances where data is missing continuously for the user input days. 

```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.backends.backend_pdf import PdfPages
import seaborn as sns
sns.set_style('whitegrid')
import matplotlib as mpl
mpl.rcParams.update({'figure.max_open_warning': 0})
%matplotlib inline
plt.tight_layout
```

```python
def categorical_dist(df,
                    brand_col,
                    granularity_col, # granularity column e.g. SKU
                    units_col,      # units sold column
                    year_col,         # year column
                    period_level): # month-wise, quarter-wise, or week-wise
    """
    Function to get categorical distribution plot and summary. Plot gives number of SKU/PPG over the period(week/month/quarter).
    Summary gives granularity and number of new items in next period and missing items from previous period.
    
        Parameters in order:
            df: Dataframe
            brand_col: Corporate segment column. Category from which SKU/PPG is plotted.
            granularity_col: SKU/PPG column for which plot is required
            units_col: Units sold column
            year_col: Year column
            period_level: Period on which distribution is required. (Month, Week, or Quarter)
        
        Returns:
            Plot of Number of items in category over period
            Dataframe with period level summary
    
    """
    
    dv = df[df[units_col] >0]
    
    dv = dv.groupby([brand_col,year_col,period_level])[granularity_col].unique().reset_index()
   
    dv['No_of_Unique_Items'] = dv[granularity_col].str.len() # getting number of unique APNs sold on particular month
   
    dv['Item_lag'] = dv[granularity_col].shift(1) # previous month APN in new column
    
    dv['Item_lag'].iloc[0] = dv[granularity_col].iloc[0]
    
    # getting APN not present in current month but present in previous month
    dv['Absent_items'] = dv['Item_lag'].map(set) - dv[granularity_col].map(set) 
     
    # getting APN present in current month but not present in previous month
    dv['New_items'] = dv[granularity_col].map(set) - dv['Item_lag'].map(set)
    
    dv.drop(['Item_lag'],axis=1,inplace=True)
    
    dv['Period'] = dv[year_col].astype(str) + "-" + dv[period_level].astype(str)
    
    # plotting at period level
    sns.set(rc={'figure.figsize':(15, 4)})
    fig, ax = plt.subplots()
    ax = sns.lineplot(data=dv, y='No_of_Unique_Items', x='Period', hue=brand_col)
    plt.xticks(rotation=30,horizontalalignment='right')
    plt.gcf().subplots_adjust(bottom=0.2)
    None
    
    # exporting
    fig.savefig('../Output/EDA/Categorical_distribution_plot.png')
    
    dv.to_csv('../Output/EDA/Categorical_distribution_Period_level_summary.csv',index=False)
    
    return dv # returns a dataset with period level summary

    
    
```

```python
def chron(df,
          date,
          granularity, 
          unit,
          start_date,
          end_date,
          misslist = [1], 
          zero_negative_sales=True): 
    
    """
    Function to find chronic missing instances or find total missing days between given start and end dates.
    To find chronic missing, pass the list of missing days in misslist parameter.
    To find total missing days, the misslist parameter can be left.
    
        Parameters in order:
            df : Dataframe
            date : Date column in datetime64 yyyy-mm-dd format
            granularity (list): Granularity list e.g. ['Segment','Retailer','APN'] for Segment X Retailer X APN
            unit : Units sold column
            start_date (yyyy-mm-dd): Start date for consideration
            end_date (yyyy-mm-dd): End date for consideration
            misslist (list)(optional): List of missing days required. e.g. [3,8,11]. Default= [1]
            zero_negative_sales (boolean): Boolean value if 0 or negative units sold be filtered. Default= True
            
        Returns:
            Dataframe with granularity and number of missing instances and total missing days based on user input.
    """
    df[date] = pd.to_datetime(df[date])
    df = df[(df[date] >= start_date) & (df[date] <= end_date)]
    
    group = [date]
    group.extend(granularity)
    dg = pd.DataFrame(df.groupby(group)[unit].sum()).reset_index()
    
    if (zero_negative_sales == True):
        dg = dg[dg[unit]>0]

    gran1 = granularity.copy()
    gran1.append(date)
    dg = dg.sort_values(gran1)
    dg['Next_date'] = dg.groupby(granularity)[date].shift(-1)
    dg['Diff'] = dg['Next_date'] - dg[date] 
    
    # first dataset to get total missing days for each granularity
    d_tot = dg.groupby(granularity).agg(Total_Missing_days = ('Diff',sum)).reset_index()
    d_tot['# of consecutive missing days'] = 'Total missing days'
    d_tot = d_tot.rename(columns = {'Total_Missing_days': '# of missing instances'})
    gran1 = granularity.copy()
    gran1.extend(['# of consecutive missing days','# of missing instances'])
    d_tot = d_tot[gran1]

    # second dataset to get missing instances for particular days
    gran1 = granularity.copy()
    gran1.append('Diff')
    dx = dg.groupby(gran1).size().reset_index(name = 'count')
    dx['y'] = 0
    dx['day'] = dx['Diff'].dt.days
    for item in misslist:
        for index,row in dx.iterrows():
            if (row['day'] > item):
                dx.loc[index,'y'] = item+1
            elif (row['day'] == item):
                dx.loc[index,'y'] = item

    gran1 = granularity.copy()
    gran1.append('y')
    dx = dx.groupby(gran1)['count'].sum().reset_index(name = '# of missing instances')

    # modifying values to get a table
    dx = dx[dx['y']>0]
    dx['y1'] = (dx['y']).apply(str)
    dx['y2'] = dx.groupby(granularity)['y1'].shift(-1)
    dx['y3'] = np.where(dx['y'].isin(misslist),dx['y'],dx['y1']+"-"+dx['y2'])
    dx['y3'] = np.where(dx['y3'].isnull(),dx['y'].astype(str)+"+",dx['y3'])

    dx = dx.rename(columns = {'y3':'# of consecutive missing days'})
    dx.drop(['y','y1','y2'],axis=1,inplace=True)
    gran1 = granularity.copy()
    gran1.extend(['# of consecutive missing days','# of missing instances'])
    dx = dx[gran1]
    
    # appending both dataframes and sorting
    dx = dx.append(d_tot,ignore_index=True)
    dx = dx.sort_values(granularity,ignore_index=True)
    
    # exporting
    dx.to_csv('../Output/EDA/Chronic_missing_and_total_missing_days.csv',index=False)

    return dx

```
